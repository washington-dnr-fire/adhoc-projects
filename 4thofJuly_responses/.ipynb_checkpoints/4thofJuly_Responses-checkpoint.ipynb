# Independence Day Responses

The 4th of July (and the days surrounding it) experiences the most ignitions of any period during fire season. These ignitions are primarily fireworks and campfires relating to human recreation. It would be interesting to verify this and see what shakes out.

**1. Import helpful libraries**

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 600
```

**2. Read in dataset from EIRS**

```python
fires_df = pd.read_csv('july4_responses.csv')
fires_df.head(2)
```

**3. Ensure we're working with datetime objects for START_DT**

```python
fires_df['START_DT'] = pd.to_datetime(fires_df['START_DT'])
```

```python
fires_df['MO_DAY'] = fires_df['START_DT'].dt.strftime('%m-%d')
fires_df['YEAR'] = fires_df['START_DT'].dt.strftime('%y')
```

**4. Only get responses that occur on 7/3, 7/4, and 7/5, then drop all the extra baggage**

```python
fires_df = fires_df[(fires_df['START_DT'].dt.strftime('%m-%d') >= '07-03') & (fires_df['START_DT'].dt.strftime('%m-%d') <= '07-05')]
fires_df = fires_df.drop(columns=['OBJECTID', 'FIREEVENT_ID', 'FIRE_TWP_WHOLE_NO', 'FIRE_TWP_FRACT_NO', 'FIRE_RGE_WHOLE_NO', 'SECTION_SUBDIV_PTS_ID', 
                      'RES_ORDER_NO', 'NON_DNR_RES_ORDER_NO', 'FIRE_RGE_FRACT_NO', 'FIRE_RGE_DIR_FLG', 'FIRE_SECT_NO', 'BURN_NONSTOCK_AREA',
                                 'BURNESCAPE_RSN_LABEL_NM', 'BURN_REPROD_AREA', 'BURN_MERCH_AREA'])
fires_df.head(2)
```

**5. What is the primary cause of fires during this period?**

```python
fires_df.groupby([('FIREGCAUSE_LABEL_NM')]).size()
```

**6. Where do we see the most responses during this period?** 

```python
ax = fires_df['REGION_NAME'].value_counts().sort_values(ascending=True).plot.barh(figsize=(8,8), width=1.0, edgecolor='black', stacked=True)
ax.set_xlabel("Number of Responses")
```

**7. In what counties do we see the most responses?**

```python
ax = fires_df['COUNTY_LABEL_NM'].value_counts().nlargest(10).sort_values(ascending=True).plot.barh(figsize=(8,8), width=1.0, edgecolor='black', stacked=True)
ax.set_xlabel("Number of Response")
```

**8. Given this trend, NE Region would like to staff a second shift during this period. When do we see responses throughout the day?**

```python
nes_df = fires_df[fires_df['REGION_NAME'] == 'NORTHEAST']
nes_df.shape
```

```python
ax1 = nes_df['START_TM'].hist(bins=24, grid=False, edgecolor='black')
ax1.set_xlabel("Time of Day (24-hr Time)")
ax1.set_ylabel("Frequency of Fires")
ax1.set_title('NE Region - Response Start Times, July 3-5, 2010-2019')
```
